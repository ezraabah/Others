Relational plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
sns.set(style='darkgrid')
```

```python
sns.relplot(x= "Kms_Driven", y= "Selling_Price", data=df)
```

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df)
```

We can also separate using hue. Hue = "separating_column".

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, hue="Fuel_Type")
```

We can also separate using style. style = "another/same_separating_column".

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, hue="Fuel_Type", style="Transmission")
```

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, style="Transmission")
```

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, hue="Year")
```

We can change the colors to something desired using palette:
Available Palettes: deep, muted, bright, pastel, dark, colorblind, RdBu <br>
We can also have Blues, Reds, greens, Yellows, etc<br>
We can also have Blues_r or Blues_d

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, hue="Fuel_Type", palette="Reds_r")
```

We can also make use of the sizes

```python
sns.relplot(x= "Present_Price", y= "Selling_Price", data=df, size="Fuel_Type",  palette="Reds")
```

```python
from numpy.random import randn
```

```python
df2 = pd.DataFrame(dict(time=np.arange(500),value=randn(500).cumsum()))
```

```python
df2.head()
```

```python
sns.relplot(x="time", y="value", data=df2)
```

To make it a time graph, We can use kind=line

```python
sns.relplot(x="time", y="value", data=df2, kind="line")
```

If x wasnt already sorted

```python
df2=pd.DataFrame(randn(500,2).cumsum(axis=0),columns=['time','value'])
```

```python
sns.relplot(x="time", y="value", data=df2, kind="line")
```

```python
sns.relplot(x="time", y="value", data=df2, kind="line", sort=False)
```

```python
sns.relplot(x="time", y="value", data=df2, kind="line", sort=True)
```
